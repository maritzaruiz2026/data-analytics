## Lab 2 of Exercise 2.B: Working with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
print(f'Today is {today.strftime("%A")}, {today.strftime("%B")} {today.strftime("%d")}, {today.strftime("%Y")}')
```

# My birthday

```python
DOB = datetime.date(1995, 12, 24)
```

```python
print(f'My birthday is {DOB.strftime("%m")}/{DOB.strftime("%d")}/{DOB.strftime("%Y")}')
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print(f'90 days from today is {ninety_d.strftime("%B")} {ninety_d.strftime("%d")}, {ninety_d.strftime("%Y")}')
```

# Time for Dinner

```python
dinner_time = datetime.time(15, 00, 00)
```

```python
print(f"Let's meet for dinner at {dinner_time.strftime("%I")}:{dinner_time.strftime("%M")} PM")
```
