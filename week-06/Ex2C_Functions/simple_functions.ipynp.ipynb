## Lab 1 of Exercise 2.C: Working with defining custom functions

```python
def favorite_things(name, movie, music):
    return(f'{name}\n{movie}\n{music}')
print(favorite_things('Maritza Ruiz', 'The Nightmare Before Christmas', 'SOAD'))
```

```python
def why_im_here(purpose):
    return(f'{purpose}')
print(why_im_here('I  joined YUU to build the skills, experience and get my foot in the door to start my career in data analytics.'))
```

```python
def favorite_place(place, why):
    return(f'One of my favorite places is {place} because {why}.')
print(favorite_place('the gym', 'I am able to de-stress, dedicate time to myself and work on my mental/physical health'))
```

```python
print(favorite_place('my home', 'I am able to recharge after life has been life-ing'))
```
