## Lab 3 of Exercise 2.C: Working with defining custom tax functions

```python
def get_soc_sec_tax(gross_pay):
    ss_tax = gross_pay * 0.062
    return ss_tax
```

```python
def get_medicare_tax(gross_pay):
    medicare_tax = gross_pay * 0.0145
    return medicare_tax
```

```python
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        federal_tax = 0.23 * gross_pay
    elif withholding_code == 1:
        federal_tax = 0.21 * gross_pay
    elif withholding_code == 2:
        federal_tax = 0.195 * gross_pay
    elif withholding_code == 3:
        federal_tax = 0.185 * gross_pay
    else:
        federal_tax = (0.18 - ((withholding_code - 4) *.005)) * gross_pay 
     
    return federal_tax
```

```python
gross_pay = 750
withholding_code = 0

federal_tax = get_federal_tax(gross_pay, withholding_code)
print(f"Person1:   gross pay: ${gross_pay}   witholding code: {withholding_code}   federal tax: ${federal_tax:.2f}")
```
