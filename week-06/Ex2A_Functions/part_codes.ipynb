# Exercise 2.A Using built-in functions 
## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(partcode):
    name = partcode.split(':')
    return name[0]
```

```python
def get_part_num(partcode):
    start = partcode.split(':')
    end = start[1].split('-')
    num = end[0]
    return num
```

```python
def get_size(partcode):
    size = partcode.split('-')
    return size[1]
```

```python
prod1 = 'ACME:123-L'
prod2 = 'DI:12345-M'
prod3 = 'ACE:1-12'
```

```python
print(f"For product {prod1} the supplier is {get_supplier_code(prod1)}. Part number {get_part_num(prod1)} is available in size {get_size(prod1)}.")
```

```python
print(f"For product {prod2} the supplier is {get_supplier_code(prod2)}. Part number {get_part_num(prod2)} is available in size {get_size(prod2)}.")
```

```python
print(f"For product {prod3} the supplier is {get_supplier_code(prod3)}. Part number {get_part_num(prod3)} is available in size {get_size(prod3)}.")
```
