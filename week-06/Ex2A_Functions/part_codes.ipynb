## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(partcode):
    name = partcode.split(':')
    return name[0]
```

```python
def get_supplier_code(partcode):
    start = partcode.split(':')
    end = start[1].split('-')
    code = end[0]
    return code
```

```python
def get_supplier_code(partcode):
    size = partcode.split('-')
    return size[1]
```

```python
prod1 = 'ACME:123-L'
prod2 = 'DI:12345-M'
prod3 = 'ACE:1-12'
```

```python
supplier = prod1.split(':')[0]
part = prod1.split(':')[1].split('-')[0]
size = prod1.split('-')[1]

print(f"For product {prod1} the supplier is {supplier}. Part number {part} is available in size {size}.")
```
